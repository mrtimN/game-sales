```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('data/regional_sales_by_genre.csv')
df = df.drop(index=[0])
df
```

```python
# Set figure size
plt.figure(figsize=(12, 6))

# Define categories (game genres)
genres = df["name"].dropna()

# Define sales data
na_sales = df["na_sales"].dropna()
eu_sales = df["eu_sales"].dropna()
jp_sales = df["jp_sales"].dropna()
other_sales = df["other_sales"].dropna()

# Define bar width
bar_width = 0.8

# Stacked bar chart
plt.bar(genres, na_sales, label="NA Sales", color="#1f77b4")
plt.bar(genres, eu_sales, bottom=na_sales, label="EU Sales", color="#ff7f0e")
plt.bar(genres, jp_sales, bottom=na_sales + eu_sales, label="JP Sales", color="#2ca02c")
plt.bar(genres, other_sales, bottom=na_sales + eu_sales + jp_sales, label="Other Sales", color="#d62728")

# Labels and title
plt.xlabel("Game Genre")
plt.ylabel("Sales (Millions)")
plt.title("Game Sales by Genre and Region")
plt.xticks(rotation=45, ha="right")  # Rotate x-axis labels
plt.legend()
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Show the plot
plt.tight_layout()
plt.savefig('graphs/regional_sales_by_genre.png', transparent=True)
plt.show()
```
